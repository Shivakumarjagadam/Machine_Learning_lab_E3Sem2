What is label Encoding?

Yes, based on the source material, here's an explanation of Label Encoding and One-Hot Encoding:
Label Encoding
Label Encoding is a method used to convert categorical (text) data into numerical form. It works by assigning a unique integer to each distinct category within a column.

The source highlights the following reasons for needing Label Encoding:

Machine Learning Requires Numeric Data: Many machine learning models, such as Linear Regression, Decision Trees, and Neural Networks, cannot directly process text data. Label encoding transforms categorical data into a numerical format that these models can understand.

Reduces Memory Usage: Storing numerical data (integers) is generally more memory-efficient than storing text.
